```python
from pymongo import MongoClient
import pprint
```

```python
client = MongoClient('mongo', 27017, username='root', password='example')
```

```python
db_meteo = client.meteo
postmeteo = db_meteo.posts
```

```python
pprint.pprint(postmeteo.find_one())
```

```python
for post in postmeteo.find():
    pprint.pprint(post)
```

```python
pprint.pprint(postmeteo.find_one({"name":"Neustadt"}))
```
